# **3. Data preparation**

## Transform strings to numerical values

Create a funcition

```python
def time2num(x):
    """
    This function transform Time (h:m:s) string to numerical value in hours.
    input: str
    output: float
    """
    if type(x) == str: # only works when x is a string
        time = x.split(':') # split at the ":", resulting in 3 parts indexed at 0, 1, 2
        # print(time[0], time[1], time[2])
        x = float(time[0]) + float(time[1])/60 + float(time[2])/3600  # convert to decimal hours
    return x
```

Test the function

```python
time2num("20:28:27")
```

Apply the function **elementwise** to the *Time (h:m:s)* column using map()
https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html

```python
df["Time (h:m:s)"].map(time2num)
```

Add the result at **Series** data type to **DataFrame**

```python
df['time_h'] = df["Time (h:m:s)"].map(time2num)
df
```

## **Model comparison (Advanced)**

Create a function to run models and compare their metrics

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def run_models(models, X_train, y_train, X_test, y_test):
    """
    input: models is a dictionary of model instances
    output: a DataFrame with train and test metrics for all models
    """
    # create an empty list to store the metrics of all models
    models_metrics=[]

    # iterate the models in the input models dictionary
    for key, model in models.items(): # must use .items()
        # The StandardScaler() to ensure the data is on the same scale, reduce the impact of outliers and improve model performance
        steps = [('scaler', StandardScaler()),
                 (str(key), model)]
        pipe = Pipeline(steps) # create a model named as pipe

        pipe.fit(X_train, y_train) # train this model
        y_train_pred = pipe.predict(X_train) # use the trained model
        y_test_pred = pipe.predict(X_test)

        # Analysze the model performance
        model_res = [] # create an empty list to store the metrics of a model

        # cross valication to reduce the overfitting
        model_res.append(round(abs(cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')).mean(), 3)) # neg: sklearn's convention of the higher the better
        model_res.append(round(abs(cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_squared_error')).mean(), 3))
        model_res.append(round(cross_val_score(pipe, X_train, y_train, cv=5, scoring='r2').mean(), 3))

        model_res.append(round(mean_absolute_error(y_test, y_test_pred), 3))
        model_res.append(round(mean_squared_error(y_test, y_test_pred), 3))
        model_res.append(round(r2_score(y_test, y_test_pred), 3))

        # append the model metrics to the models_metrics list
        models_metrics.append(model_res)

    indx = [key for key in models.keys()] #intent to get the keys
    table = pd.DataFrame(models_metrics,
                         columns = ['train_MAE', 'train_MSE', 'train_R2', 'test_MAE', 'test_MSE', 'test_R2',],
                         index = indx)
    return table
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

# create model instances
models = {'lr': LinearRegression(),
          'knn': KNeighborsRegressor(),
          'dtr': DecisionTreeRegressor(),
          'rfr': RandomForestRegressor(),
          'xgbr': XGBRegressor()}
```

```python
# run the model comparason function
run_models(models, X_train, y_train, X_test, y_test)
```
